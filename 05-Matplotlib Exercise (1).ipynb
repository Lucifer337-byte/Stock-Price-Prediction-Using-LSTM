## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
import matplotlib.pyplot as plt 
fig = plt.figure() 
ax1 = fig.add_axes([0, 0, 1, 1]) 
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) 
x 
y  
ax1.plot(x, y)
ax1.set_xlabel('X Axis Label (ax1)')
ax1.set_ylabel('Y Axis Label (ax1)')
ax1.set_title('Main Plot (ax1)')
ax2.plot(x, y)
ax2.set_xlabel('X Axis Label (ax2)')
ax2.set_ylabel('Y Axis Label (ax2)')
ax2.set_title('Inset Plot (ax2)')
plt.show()
```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.plot(x, y, label="y = 2x", color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Plot of y = 2x on ax1')
ax1.legend()
plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.plot(x, y, label="y = 2x", color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Plot of y = 2x on ax1')
ax1.legend()
ax2.plot(x, z, label="z = x^2", color='red')
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_title('Plot of z = x^2 on ax2')
ax2.legend()
plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, (ax1, ax2) = plt.subplots(1, 2) 
ax1.plot(x, y, 'b--', linewidth=2)   
ax1.set_title('Linear') 
ax2.plot(x, z, 'r-', linewidth=3)   
ax2.set_title('Quadratic')
plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))   
ax1.plot(x, y, 'b-', linewidth=2)  
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x, z, 'r--', linewidth=2)  
ax2.set_xlabel('x')
ax2.set_ylabel('z')
plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))   
ax1.plot(x, y, 'b--', linewidth=2)   
ax1.set_title('Linear')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x, z, 'r-', linewidth=3) 
ax2.set_title('Quadratic')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
plt.tight_layout()  
plt.show()
```
